```python
import pymongo
from pymongo import MongoClient
client = MongoClient('mongodb://broker:broker@dev-2.fugle.tw:27017')
db = client['company']
collection = db['revenue']
```

```python
all_data = list(collection.find({"year":{"$gte":2014},"symbol_id":"2317"},#))
                     {'_id': 0,'symbol_id': 1,'year':1,'month':1,'data':1}))
all_data
```

```python
revenue = []
for item in all_data:
    data = item['data']['current']['revenue']
    revenue.append(data)
```

```python
yoy = []
for item in all_data:
    data = item['data']['current']['yoy']
    yoy.append(data)
```

```python
year = []
month = []
symbol_id = []
for item in all_data:
    year_ = item['year']
    month_ = item['month']
    id_ = item['symbol_id']
    year.append(year_)
    month.append(month_)
    symbol_id.append(id_)
```

```python
import pandas as pd
df = pd.DataFrame({"year":year,
                   "month":month,
                   "symbol_id":symbol_id,
                   "yoy":yoy,
                   "revenue":revenue})
df
```

```python
df.to_csv("yoy_revenue_cuurent_2317.csv")
```

```python
import pymongo
from pymongo import MongoClient
client = MongoClient('mongodb://broker:broker@dev-2.fugle.tw:27017')
db = client['institutional']
collection = db['tdcc_distribution']
```

```python
import datetime
all_data = list(collection.find({"date":{"$gte":datetime.datetime(2014, 1, 1, 0, 0, 0, 0)},"symbol_id":"2317"},#))
                     {'_id': 0,'symbol_id': 1,'date':1,'data':1}))
all_data
```

```python
value = []
for i in range(len(all_data)):
    data = all_data[i]['data']
    for n in range(len(data)):
        new = data[n]
        if new['range_string'] == '合計':
            #print(code_)
            value.append(new['shares'])
            #print(all_data[i]['date'])
print(len(value))
```

```python
date = []
for item in all_data:
    date.append(item['date'])
print(len(date))
```

```python
df = pd.DataFrame({"date":date,
                   "shares":value,
                   "id":"2317"})
df
```

```python
df.to_csv("shares_2317.csv")
```

```python
import pymongo
from pymongo import MongoClient
client = MongoClient('mongodb://broker:broker@dev-2.fugle.tw:27017')
db = client['technical']
collection = db['tw_stock_price']
```

```python
all_data = list(collection.find({"year":{"$gte":2015},"symbol_id":"2317"},#))
                     {'_id': 0,'history':1}))
all_data
```

```python
date = []
close = []
for n in range(len(all_data)):
    for nn in range(len(all_data[n]['history'])):
        date.append(all_data[n]['history'][nn]['date']+datetime.timedelta(hours=8))
        close.append(all_data[n]['history'][nn]['close'])
```

```python
df = pd.DataFrame({"date":date,
                   "close":close,
                   "id":"2317"})
df
```

```python
df.to_csv("price_2317.csv")
```
