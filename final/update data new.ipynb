```python
def trans_time(x):
    #convert date from str to datetime
    x = datetime.datetime.strptime(x,'%Y-%m-%dT%H:%M:%S.%fZ')

    return x
```

### 最後收盤價

```python
price_col = client['technical']['tw_stock_price']
year = datetime.datetime.today().year

def query_latest_close(symbol_data):

    try:
        query = price_col.find({'symbol_id': symbol_data['symbol_id'],
                                'year': year},
                               {'_id':0, 'history':1})

        last_data = list(query)[0]['history'][-2]
        symbol_data['last_close'] = last_data['close']
        
        ##trans time
        symbol_data['last_date'] = last_data['date']
        symbol_data['date'] = trans_time(symbol_data['date'])
        
        ##volume 股 to 張
        symbol_data['volume'] = int(symbol_data['volume'] / 1000)
        
        return symbol_data
    
    except IndexError:
        pass
```

### 最新現金股利

```python
def query_latest_div(symbol_data):
    
    try:
        a = requests.get(
            'https://www.fugle.tw/api/v1/data/new_content/FCNT000074?symbol_id='+
            symbol_data['symbol_id'])

        msg = a.json()['rawContent']

        if msg['div_year'] >= 2018:
            div_rate = (msg['cash_div'] / symbol_data['close'])*100
            div_rate = round(div_rate, 2)
            symbol_data['div_rate'] = div_rate
        else:
            symbol_data['div_rate'] = 0
        
    except KeyError:
        symbol_data['div_rate'] = 0
        
    return symbol_data
```

### 近兩週集保資料

```python
def insti_growth_rate(now_data, last_data, value):

    def get_growth_rate(data, value):

        range_shares = list(filter(lambda x: type(x['range_start'])==int and x['range_start']>=value, data))

        #set a function to sum shares, people and proportion data
        add_func = lambda x: lambda y, z: y + z[x]

        sum_shares = reduce(add_func('shares'), range_shares,0) / 1000 #calculate sum of shares
        sum_people = reduce(add_func('people'), range_shares,0) #calculate sum of people
        
        try:
            avghold =  sum_shares / sum_people #calculate average shareholding per person
        except ZeroDivisionError:
            avghold = 0
        proportion = reduce(add_func('proportion'), range_shares,0) #calculate sum of proportion

        return avghold, proportion

    now_avghold, now_proportion = get_growth_rate(now_data,value) #get this week average shareholding and proportion data
    last_avghold, last_proportion = get_growth_rate(last_data,value) #get last week average shareholding and proportion data
    
    try:
        avghold_WoW = ((now_avghold-last_avghold)/last_avghold)*100 #calculate the growth rate of average sharesholding
        avghold_WoW = round(avghold_WoW, 3)
        
    except ZeroDivisionError:
        avghold_WoW = None
        
    proportion_WoW = now_proportion - last_proportion #calculate the growth rate of proportion
    proportion_WoW = round(proportion_WoW, 3)

    return avghold_WoW, proportion_WoW


def query_latest_insti(symbol_data):

    tdcc_query = requests.get(
        'https://www.fugle.tw/api/v1/data/new_content/FCNT000022?symbol_id='+
        symbol_data['symbol_id'])
    
    now, last = tdcc_query.json()['rawContent'][:2] #取最新兩周
    
    avghold_WoW, proportion_WoW = insti_growth_rate(now['data'],
                                                    last['data'],
                                                    1000001)
    symbol_data['avghold_wow'] = avghold_WoW
    symbol_data['proportion_wow'] = proportion_WoW
    
    return symbol_data
```

### 將 change div insti 資料之key缺失職篩選掉

```python
def filter_change(data):
    
    change_data = list(filter(lambda x: x['change_rate'] 
                              if 'change_rate' in x else None, data['symbols']))
    
    data['symbols'] = list(filter(lambda x: query_latest_close(x), change_data))
    
    #print(data['symbols'][0]['last_date'], data['symbols'][0]['date'])
    
    return data

def filter_div(data):
    
    div_data = list(filter(lambda x: query_latest_div(x), data['symbols']))
    
    data['symbols'] = div_data
    
    return data
    
def filter_insti(data):
    
    insti_data = list(filter(lambda x: query_latest_insti(x), data['symbols']))
    
    data['symbols'] = insti_data
    
    return data
```

```python
def equal_weight_change(symbols):
    
    change_num = len(symbols)
    change_sum = sum(x['change_rate'] for x in symbols)
    
    try:
        # equal weight change of this collection
        equal_weight_change = round(change_sum / change_num, 2)
        return equal_weight_change
        
    except ZeroDivisionError:
        pass
```

```python
def equal_volume_change(symbols):
    
    # daily return that buy one for each stock
    close_sum = sum(list(map(lambda x: x['close'], symbols)))
    last_close_sum = sum(list(map(lambda x: x['last_close'], symbols)))
    
    try:
        equal_volume_change = (close_sum / last_close_sum - 1)*100
        equal_volume_change = round(equal_volume_change, 2)

        return equal_volume_change
    except ZeroDivisionError:
        pass
```

```python
def top_3_change(symbols):
    
    sort_data = sorted(symbols, 
                       key=lambda x: x['change_rate'], 
                       reverse = True)[:3]
    
#     sort_data = [{'symbol_id': i['symbol_id'],
#                   'name': i['name'], 
#                   'change_rate': i['change_rate']} for i in sort_data]
    return sort_data
```

```python
def avg_volume(symbols):
    
    volume_num = len(symbols)
    volume_sum = sum(x['volume'] for x in symbols)
    # average volume of this collection
    try:
        avg_volume = int(volume_sum / volume_num)

        return avg_volume
    except ZeroDivisionError:
        pass
```

```python
def avg_div(symbols):
    
    div_num = len(symbols)
    div_sum = sum(x['div_rate'] for x in symbols)
    
    try:
        # average div of this collection
        avg_div = round(div_sum / div_num, 2)

        return avg_div
    except ZeroDivisionError:
        pass
```

```python
def avg_avghold(symbols):
    
    avghold_list = [x['avghold_wow'] 
                    for x in symbols if x['avghold_wow'] != None]
    
    avghold_num = len(avghold_list)
    avghold_sum = sum(avghold_list)
    
    try:
        # average avg_avghold of this collection
        avg_avghold = round(avghold_sum / avghold_num, 2)

        return avg_avghold
    except ZeroDivisionError:
        pass
```

```python
def avg_proportion(symbols):
    
    pro_num = len(symbols)
    pro_sum = sum(x['proportion_wow'] for x in symbols)
    
    try:
        # average avg_avghold of this collection
        avg_pro = round(pro_sum / pro_num, 2)

        return avg_pro
    except ZeroDivisionError:
        pass
```

```python
def sort_data(data, key, word, reverse=True):
    
    rank_num = ceil(len(data)*0.2)
    if rank_num > 10:
        rank_num = 10
    else:
        pass
    
    data = [x for x in data if x[key] != None]
    sort_data = sorted(data, key=lambda x: x[key], reverse=reverse)[:rank_num]
    
    if 'change' in key and reverse == True:
        sort_data = [x for x in sort_data if x[key] > 0]
        
    elif 'change' in key and reverse == False:
        sort_data = [x for x in sort_data if x[key] < 0]
    
    for i in range(len(sort_data)):
        print(word+' No.'+str(i+1))
        sort_data[i][key+'_rank'] = word+' No.'+str(i+1)
        
    return data
```

```python
def get_return_volume(interval, df, symbol):
    test = []
    for i in range(0, 51, 5):
        try:
            v = df.loc[trading_date[i+interval]['date']:trading_date[i]['date']]
            #news = df_news.loc[trading_date[i+interval]['timestamp']:trading_date[i]['timestamp']]
            b = news_col.find({'symbol_ids':symbol,'timestamp':{'$lte':trading_date[i]['date'],'$gt':trading_date[i+interval]['date']}},
                              {'_id':0,'symbol_ids':1,'timestamp':1}).sort('timestamp',-1).limit(1000) 
            news_data = list(b)
            
            return_ = round((v['close'].iat[-1] / v['close'].iat[0] - 1)*100, 2)
            volume_ = sum(v[1:]['volume'])
            news_ = len(news_data)
            
            test.append({
                'date': v.index[-1],
                'return': return_,
                'volume': volume_,
                'news':news_
            })

        
        except IndexError:
            continue
            
    return test
```
