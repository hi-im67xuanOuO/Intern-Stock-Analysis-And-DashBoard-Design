```python
import pprint
```

```python
pprint.pprint(collection.find_one())

# 返回的文檔包含"_id"，會在插入時自動添加
```

```python
pprint.pprint(collection.find_one({"symbol_id": "3125"}))
```

```python
# 查詢多個相同指定條件的
for item in collection.find({"industry_type": "電子工業"}):
    pprint.pprint(item)
```

```python
# 計數
#collection.count_documents({})
collection.count_documents({"industry_type" : "電子工業"})
```

```python
# 範圍查詢
d = list(map(str,range(3120,3125,1))) #轉換3120 to 3125 為文字
d
for item in collection.find({"symbol_id" : {"$in" : d}}).sort("symbol_id"):
    pprint.pprint(item)
```

```python
# 查詢多個相同指定條件的
for item in collection.find({"industry_type": "電子工業"}):
    pprint.pprint(item['name'])
```
